```python
import torch

node2vec_dict = {}

# Load node2vec embeddings
with open("embeddings/node2vec_emb.pkl", "rb") as fIn:
    node2vec_dict = torch.load(fIn)
```

```python
import pickle

embedding_dict = {}

# Load sentence transformers embedding
with open("embeddings/sentence_transformers_emb.pkl", "rb") as fIn:
        embedding_dict = pickle.load(fIn)
```

```python
import pandas as pd

# Load dataset
dataset = pd.read_csv('data/Mastodon_instances_metadata_v3.csv')
dataset.head()
```

```python
import torch

mixed_dict = {}

# Mixing the embeddings
for node in dataset['instance'].values:
    if node in node2vec_dict and node in embedding_dict:
        n2v_embedding = node2vec_dict[node]
        st_embedding = embedding_dict[node]

        mixed_dict[node] = torch.add(torch.tensor(n2v_embedding), torch.tensor(st_embedding)) / 2
```

```python
# Saving the mixed embedding dictionary
embedding_cache_path = "embeddings/mixed_emb.pkl"
torch.save(mixed_dict, embedding_cache_path)
```
